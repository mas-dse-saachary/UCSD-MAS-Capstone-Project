```python
def frame_to_list(x):
    L = []
    for w in x:
        L.append(w)
    return L
```

```python
def frame_to_set(x):
    L = set()
    
    for w in x:
        L.add(w)
        
    return L
```

```python
listings_events_2018_05_22_V1[['listing_id', 'event_title', 'event_id', 'event_type']].groupby(['listing_id', 'event_title'])['event_type'].apply(frame_to_set)
```

```python
def set_to_value(x):
    L = []
    
    for w in x:
        L.append(w)
        
    return L[0]
```

```python
listings_events_2018_05_22_V1[['listing_id', 'event_title', 'event_id', 'event_type']].groupby(['listing_id', 'event_title'])['event_type'].apply(frame_to_set).apply(set_to_value)
```
